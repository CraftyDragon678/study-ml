```python
import tensorflow.keras as keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(units=1, input_dim=3))
# model.add(keras.layers.Activation('linear'))

model.compile(loss='mse', optimizer=keras.optimizers.SGD(lr=1e-5))
model.summary()
history = model.fit(x_data, y_data, epochs=100)
```

```python
# plt.plot(x_data, y_data, 'o')
# plt.plot([72., 93., 90.], model.predict(np.array([[72., 93., 90.]])))
model.predict(np.array([[72., 93., 90.], [72., 93., 90.], [72., 93., 90.]]))
```
