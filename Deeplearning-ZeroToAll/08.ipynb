# 궁극의 목표

인공지능이 스스로 생각하게 만드는 것이 목표이다.

그래서 스스로 생각하는 사람 뇌 구조를 보니깐 뉴럴 이라는 것이 있는데
굉장히 단순하게 동작하고 있었다

이것을 기계적으로 만든 것이 Activation Function이다.

입력을 받아서 곱하고(weight) 더해서(bias)
일정 값이 넘어가면 신호를 주고(1) 아니면 주지 않는다(0)


But.. 문제가 생겼다.

# XOR Problem
or 이나 and는 다음과 같이 linear한 모습을 기대할 수 있다.

|or|0|1|
|-|-|-|
|0|-|+|
|1|+|+|

|and|0|1|
|-|-|-|
|0|-|-|
|1|-|+|

그런데 xor은?

|xor|0|1|
|-|-|-|
|0|-|+|
|1|+|-|

linear한 모습이 나오지 않는다.

심지어 이것은 수학적으로도 불가능하다고 증명이 되었다.
> Perceptrons(1969): 여러개를 합치면 불가능하다. 하지만 학습은 어떻게..?
