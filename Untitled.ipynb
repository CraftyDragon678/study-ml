```python
import matplotlib.pyplot as plt
import numpy as np

num_points = 100
vectors_set = []
for i in range(num_points):
    x1 = np.random.normal(0, 3)
    y1 = x1 * 0.4 + np.random.normal(0, 1)
    vectors_set.append([x1, y1])

x_data = [v[0] for v in vectors_set]
y_data = [v[1] for v in vectors_set]

plt.plot(x_data, y_data, 'o', markersize=3)
plt.plot([-20, 20], [-20 * 0.4, 20 * 0.4], linewidth=2)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True, which='major')
```

```python
import numpy as np

x_data = np.array([1, 2, 3])
y_data = np.array([1, 2, 3])
xs = []
ys = []
for W in np.arange(-3, 5, 0.01):
    xs.append(W)
    ys.append(np.mean(np.square(W * x_data - y_data)))
plt.plot(xs, ys)
plt.xlabel("W")
plt.ylabel("cost")
```

```python
from matplotlib import cm
fig = plt.figure()
ax = fig.gca(projection='3d')

x = np.arange(-4, 4, 0.1)
y = np.arange(-4, 4, 0.1)
x, y = np.meshgrid(x, y)
z = x**2 + 0.3 * x * y + y**2
ax.plot_surface(x, y, z, cmap=cm.rainbow)
```
